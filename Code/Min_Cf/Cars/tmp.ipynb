```python
X_vals = [0.0, 0.19, 0.21, 0.40, 0.50]
Y_vals = [0.0, 0.30, 0.45, 0.55, 0.80]
q_x, q_y = (0.20, 0.50)

points = []
for x in X_vals:
    for y in Y_vals:
        dist = abs(x - q_x) + abs(y - q_y)  # L1
        points.append(((x,y), dist))

# Sort by distance ascending
points.sort(key=lambda p: p[1])

print("All points sorted by L1 distance:")
for i, (pt, d) in enumerate(points, start=1):
    print(f"{i}) {pt}, dist={d:.3f}")
```
