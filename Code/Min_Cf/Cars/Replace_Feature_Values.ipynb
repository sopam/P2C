```python
import pandas as pd

def replace_symbols(df):
    """
    Replaces '<' with 'less_than' and '>' with 'greater_than' in all DataFrame values.
    
    Args:
    df (pd.DataFrame): Input DataFrame
    
    Returns:
    pd.DataFrame: Modified DataFrame with replaced values
    """
    return df.applymap(lambda x: str(x).replace('=', 'equals').replace('<', 'less_than').replace('>', 'greater_than').replace(' ', '_') if isinstance(x, str) else x)

```

```python
train_path = './Original_Data/train_fold_1.csv'
#test_path = './Original_Data/test_fold_1.csv'
#test_path = './Data/test_fold_1.csv'
```

```python
#df = pd.read_csv(train_path)
df = pd.read_csv(train_path)

df
```

```python
# Replace '<' and '>'
df_cleaned = replace_symbols(df)

# Display the modified DataFrame
df_cleaned
```

```python
df_cleaned.to_csv('Data/train_fold_1.csv')
```
