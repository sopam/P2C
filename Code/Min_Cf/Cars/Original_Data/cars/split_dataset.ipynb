```python
path = 'cars.csv'
```

```python
df = pd.read_csv(path)
```

```python
df = df.drop('id', axis=1)
df
```

```python
# Method 3: Using map
df.columns = list(map(lambda x: x[0].upper() + x[1:], df.columns))

df
```

```python
# Split data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
X_train.to_csv('train_fold_1.csv', index = False)
X_test.to_csv('test_fold_1.csv', index = False)
```
